### Pandas Task Json
#### Q. To split 'user' column dictionary, separate and store them into different columns and store it to csv file

##### Importing Libraires

```python
import pandas as pd
from ast import literal_eval
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
raw_data = requests.get(url)
raw_data
```

```python
data = raw_data.json()
```

##### Number of dictionary in data

```python
len(data)
```

##### Preview first dictionary

```python
data[0]
```

```python
df = pd.DataFrame(data, columns=['id', 'url', 'repository_url', 'labels_url', 'user'])
df.head()
```

##### Code to Create Seperate Columns

```python
for k in df['user'].iloc[0]:
    lst = []
    for i in range(len(df)):
        lst.append(df['user'].iloc[i][k])
        
    if k =='id':
        df['user-id'] = lst  # Renaming the id to user-id so that we dont overwrite the existing id column
    elif k =='url':
        df['user-url'] = lst # Renaming the id to user-ur so that we dont overwrite the existing url column
        
    df[k] = lst  # Inserting all the values to respective column
```
