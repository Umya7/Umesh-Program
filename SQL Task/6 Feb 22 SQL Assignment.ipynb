# This function will dump data into table

```python
import csv
with open('car.data','r') as f:
    car_data=csv.reader(f,delimiter='\n')
    for i in car_data:
        lst = i[0].split(',')
        #print(type(i[0]))
        #print(f'insert into car values ({str(i[0])})')
        cursor.execute("insert into cardataset.car values('{0}','{1}','{2}','{3}','{4}','{5}','{6}')".format(lst[0],lst[1],lst[2],lst[3],lst[4],lst[5],lst[6]))
```

```python
mydb.commit()
```
