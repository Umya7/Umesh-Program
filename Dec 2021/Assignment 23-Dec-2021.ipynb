```python
def Dic_Values(d):
    
    """Function that takes input as a dictionary and gives output as the list of all the values"""
    
    for v in d.values():
        if isinstance(v, dict):
            yield from Dic_Values(v)
        else:
            yield v


a = {"a" : 5, "b" :7, "c" : {"d" : 4, "e" : 9, "f" : {"g" : 6, "k" :{"l" : 1, "m" : 3}}}}

print(list(Dic_Values(a)))
```

# Q.5 :- Write a function which will take multiple list as a input and give me concatnation of all the element as and output

```python
def listConcat(*args):
    
    """This function may take multiple lists as input and give concatenation of all the elements as output"""
    
    lst=[]
    for i in args:
        lst.extend(i)
    return lst
```

```python
listConcat([1,2,3],["umesh","prajapati","kumar"], [25,26])
```

```python
def Concat_list(*args):
    
    """This function takes n number of list as input and gives the concatnation of all the list element as one list"""
    
    l = []
    for i in args :
        if type(i) == list :
            l = l + i
    return f"Concatnation of all list element in one list is {l}"

Concat_list([1,2,3],["umesh","prajapati","kumar"], [25,26])
```

# Q.11 :- Write a function which can read video file and play for you

```python
import cv2
def PlayVideo(videofile, mirror=False):
    "This function take video file name and play for you"
    cap = cv2.VideoCapture(videofile)
    ret, frame = cap.read()
    while(1):
       ret, frame = cap.read()
       if cv2.waitKey(1) & 0xFF == ord('q') or ret==False :
           cap.release()
           cv2.destroyAllWindows()
           break
       cv2.imshow('frame',frame)
PlayVideo("video.mp4")
#to update cv palckage need to run this code in jupyter "pip install opencv-python"
```

```python
import time, vlc
 
# method to play video
def video(source):
     
    # creating a vlc instance
    vlc_instance = vlc.Instance()
     
    # creating a media player
    player = vlc_instance.media_player_new()
     
    # creating a media
    media = vlc_instance.media_new(source)
     
    # setting media to the player
    player.set_media(media)
     
    # play the video
    player.play()
     
    # wait time
    time.sleep(0.5)
     
    # getting the duration of the video
    duration = player.get_length()
     
    # printing the duration of the video
    print("Duration : " + str(duration))
     
# call the video method
video("video.mp4")
```

# Q.16 :- Write a function to read a complete PDf file

```python
import PyPDF2
def ReadPdf(path):
    
    """This functions reades the text from a pdf file from a specified folder"""
    
    pdfFileObject = open(path, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObject, strict=False)
    text=''
    for i in range(0,pdfReader.numPages):
        pageObj = pdfReader.getPage(i)
        text=text+pageObj.extractText()
    print(text)
```

# Q.17 :- Write a function to read a word file 

```python
import docx
def ReadWordFile(filename):
    
    """This function reads the word file from a specified path"""
    
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return '\n'.join(fullText)

print (ReadWordFile('C:\\Users\\upraj.LAPTOP-9SF064MV\\Documents\\The.docx'))
```

# Q.20 :- Write a function by which you will be able to append two PDF files

```python
import PyPDF2
from PyPDF2 import PdfFileMerger
def merger():
    
    """This function merges two or more pdf files lying in a same directory if the file name is provided with extenstion"""
    
    pdfs = ['The.pdf', 'The1.pdf']
    merger = PdfFileMerger()
    for pdf in pdfs:
        merger.append(pdf)
    merger.write("The6.pdf")
    merger.close()
```
