```python
#import pandas and read in the csv file containing titanic data
#show the first 5 rows of the dataset

import pandas as pd

titanic = pd.read_csv('titanic.csv')
titanic.head(5)
```

#### Q1. Find out how many male and female passenger was onboared.

```python
#total number of male passengers and total number of female passengers was onboared

total_male_female = titanic.groupby('Sex')['Sex'].count()
total_male_female
```

#### Q2. How many Survived we have ?

```python
# Create Survival Label Column
titanic['Survival'] = titanic.Survived.map({0 : 'Died', 1 : 'Survived'})
titanic.Survival.head()
```

```python
# Total Male females survived?

total_male_female_survived = titanic.groupby('Survival')['Survival'].count()[1]
total_male_female_survived
```

#### Q3. How many casuality we have?

```python
# Total Male females casuality?

total_male_female_casuality = titanic.groupby('Survival')['Survival'].count()[0]
total_male_female_casuality
```

#### Q4. What is name of a person who is the eldest one?

```python
# name of a person who is the eldest one

titanic[titanic['Age'] == titanic['Age'].max()]
```

#### Q5. How many passenger do we have in 1st, 2nd and 3rd class?

```python
# Create Pclass Label Column
titanic['Pclass'] = titanic.Pclass.map({1 : 'First Class', 2 : 'Second Class', 3 : 'Third Class'})
titanic.Pclass.head()
```

```python
titanic.groupby('Pclass')['Pclass'].count()
```

#### Q6. How many person we have whose name start with 's'?

#### Q10. From a cabin seperate text and numeric value?

```python
char = []
num = []
for i in titanic['Cabin']:
    cabin_text = ''
    cabin_number = ''
    if type(i) == str:
        for ch in i:
            if ch >= 'A' and ch <= 'Z' or ch >= 'a' and ch <= 'z':
                cabin_text += ch
            else:
                cabin_number += ch
        
        char.append(cabin_text)
        num.append(cabin_number)
        
    else:
        char.append(i)
        num.append(i)
```

```python
titanic['Cabin_char'] = char
titanic['Cabin_num'] = num
```
